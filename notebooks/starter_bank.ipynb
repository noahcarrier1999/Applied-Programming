Amount of unknown

```
print()
print("Before changing ")
print(campaign.isin(['unknown']).sum())

campaign.replace('unknown',pd.NA, inplace=True)
print()
print("Missing Values after replacing 'unknown'")
print(campaign.isnull().sum())
```

```
print("fill categorical colums with unknown")
campaign['default'].fillna('no', inplace=True)
campaign['housing'].fillna('no', inplace=True)
campaign['loan'].fillna('no', inplace=True)


print("Missing values after domain-specific imputation:")
print(campaign.isnull().sum())
```
